```python
infile = open("E:/명지대/5학년(초과학기)/3. 파이썬프로그래밍입문/13주차_파일과 객체/실습/꿈나라별나라.txt", "r")
outfile = open("E:/명지대/5학년(초과학기)/3. 파이썬프로그래밍입문/13주차_파일과 객체/실습/꿈나라별나라2.txt", "w")
```

```python
import random

fin = ["요", "자", "까", "지"]
wow = ["wow!", "좋아!", "그럴까?", "정말?", "그래."]

for line in infile:
    word_list = line.split()   # 빈칸 기준으로 단어 분리
    for word in word_list:
        print(word, end=" ")   # 화면에 출력
        outfile.write(word + " ")   # 파일에 출력
        if word[-1] in fin:   # 단어의 끝이 fin리스트 안의 글자로 끝나면
            print("  나: " + random.choice(wow))   # wow리스트 안의 글자를 랜덤으로 출력해서 붙이기
            outfile.write("  나:" + random.choice(wow)+"\n")   # 파일에 출력
        else:
            print()   # 엔터 출력
            outfile.write("\n")   # 엔터를 넣어줌
    print("-"*50)

infile.close()
outfile.close()
```
